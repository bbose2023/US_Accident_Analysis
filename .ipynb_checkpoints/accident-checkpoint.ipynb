```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# confirm that the "autosaurus" database is in MongoDB
print(mongo.list_database_names())
```

```python
# assign the database to a variable name
db = mongo['us_accidents_db']
# review the collections in our new database
print(db.list_collection_names())
```

```python
# assign each collection to a variable
accident = db['us_accident']
```

```python
accident.count_documents({})
```

```python
#Calculate total fatalities over all years
total_fatalities = accident.count_documents({})

#Calculate fatalities by year
pipeline_year = [
    {"$group": {"_id": "$YEAR", "total_fatalities": {"$sum": 1}}},
    {"$sort": {"_id": -1}}  # Sort by year descending
]
fatalities_by_year = list(accident.aggregate(pipeline_year))
df_year = pd.DataFrame(fatalities_by_year).rename(columns={'_id': 'Year'})

#Calculate fatalities by state
pipeline_state = [
    {"$group": {"_id": "$STATENAME", "total_fatalities": {"$sum": 1}}},
    {"$sort": {"total_fatalities": -1}}  # Sort by fatalities descending
]
fatalities_by_state = list(accident.aggregate(pipeline_state))
df_state = pd.DataFrame(fatalities_by_state).rename(columns={'_id': 'State'})

#Calculate fatalities by month
pipeline_month = [
    {"$group": {"_id": "$MONTHNAME", "total_fatalities": {"$sum": 1}}},
    {"$sort": {"_id": 1}}  # Sort by month name alphabetically
]
fatalities_by_month = list(accident.aggregate(pipeline_month))
df_month = pd.DataFrame(fatalities_by_month).rename(columns={'_id': 'Month'})

#Calculate fatalities by county
pipeline_county = [
    {"$group": {"_id": "$COUNTYNAME", "total_fatalities": {"$sum": 1}}},
    {"$sort": {"_id": 1}}  # Sort by county name alphabetically
]
fatalities_by_county = list(accident.aggregate(pipeline_county))
df_county = pd.DataFrame(fatalities_by_county).rename(columns={'_id': 'County'})

#Calculate fatalities by city
pipeline_city = [
    {"$group": {"_id": "$CITYNAME", "total_fatalities": {"$sum": 1}}},
    {"$sort": {"_id": 1}}  # Sort by city name alphabetically
]
fatalities_by_city = list(accident.aggregate(pipeline_city))
df_city = pd.DataFrame(fatalities_by_city).rename(columns={'_id': 'City'})

#Calculate fatalities by rural vs urban areas
pipeline_rural_urban = [
    {"$group": {"_id": "$RUR_URBNAME", "total_fatalities": {"$sum": 1}}}
]
fatalities_by_rural_urban = list(accident.aggregate(pipeline_rural_urban))
df_rural_urban = pd.DataFrame(fatalities_by_rural_urban).rename(columns={'_id': 'Rural/Urban'})

#Calculate fatalities by time of day (Daytime vs Nighttime)
pipeline_time_of_day = [
    {"$group": {"_id": "$LGT_CONDNAME", "total_fatalities": {"$sum": 1}}}
]
fatalities_by_time_of_day = list(accident.aggregate(pipeline_time_of_day))
df_time_of_day = pd.DataFrame(fatalities_by_time_of_day).rename(columns={'_id': 'Time of Day'})


#Display the data
print("Total Fatalities: ", total_fatalities)
print("\nFatalities by Year:")
print(df_year)

print("\nFatalities by State:")
print(df_state)

print("\nFatalities by Month:")
print(df_month)

print("\nFatalities by County:")
print(df_county)

print("\nFatalities by City:")
print(df_city)
      
print("\nFatalities by Rural/Urban:")
print(df_rural_urban)

print("\nFatalities by Time of Day:")
print(df_time_of_day)
```
